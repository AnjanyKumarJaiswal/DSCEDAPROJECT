##### Plot a Graph based on people highest IQ level from the given dataset.

```python
TOPIQ = df.sort_values(by='IQ', ascending=False)
name = df['Name'].to_numpy()
iq = df['IQ'].to_numpy()
plt.barh(name[0:10],iq[0:10])
plt.xlabel('IQ')
plt.ylabel('Name')
plt.title('This is highest IQ level from the datset')
plt.show()
```

##### What of the following have got highest total marks?

```python
bestmarks_temp = pd.DataFrame(columns=['Name','Physics_Marks', 'Chemistry_Marks', 'Math_Marks'])

for i in range(10):
  name = df.loc[i,'Name']
  bestmarks = (df.loc[i, 'Physics_Marks'] + df.loc[i, 'Chemistry_Marks'] + df.loc[i, 'Math_Marks']).sum()
  bestmarks_temp = pd.DataFrame({ 'Name': name,'total_marks': bestmarks}, index=[i])

bestmarks_temp = bestmarks_temp.sort_values(by=['total_marks'],ascending=False).head(10)
print(bestmarks_temp[['Name','total_marks']])
```
